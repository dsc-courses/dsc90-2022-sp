### Solutions

#### Question 1a

We need to find the third vertex of the triangle whose area is $\frac{3}{4}$ times the area between the red line and the parabola. This point is the point on the parabola whose derivative is equal to the slope of the red line.

The slope of the red line is 2, so we need to find the point on the parabola whose derivative is 2. Since the parabola has equation $f(x) = x^2$, we have $f'(x) = 2x$. Setting $2x = 2$ yields $x = 1$, meaning that the third vertex of the triangle in question is the point $(1, f(1)) = (1, 1)$.

Now we have the three vertexes of the triangle in question: $(-2, 4), (4, 16),$$ and $$(1, 1)$. We can use [this solver](https://keisan.casio.com/exec/system/1223520411) to find the area of this triangle, which is 27.

Then, the area between the red line and parabola is $\frac{4}{3}$ times the area of this triangle, which means the area between the red line and parabola is

$$\frac{4}{3} \cdot 27 = \boxed{36}$$

#### Question 1b

In the region that is enclosed between the red line and parabola, the red line is above the parabola. So, the function we need to integrate is $2x + 8 - x^2$, and we need to integrate it between $x=-2$ and $x=4$.

$$
A = \int_{-2}^4 \left( 2x + 8 - x^2 \right) dx \\
= \left[ x^2 + 8x - \frac{1}{3}x^3 \right]_{-2}^4 \\
= 16 + 32 - \frac{64}{3} - \left(4 - 16 + \frac{8}{3}\right) \\
= 48 + 12 - \frac{64 + 8}{3} \\
= 60 - 24 = \boxed{36}
$$

This matches our answer from part (a).

## Question 2

Consider the function

$$f(x) = x(48 - x^2)$$

(a) Using Fermat's method of adequality, determine the value of $x$ that maximizes $f(x)$. Is this a local or global maximum? Show your work.

(b) Using Fermat's method of adequality, determine the slope of the tangent line to $f(x)$ at $x = 1$. Verify that your answer is correct by finding $f'(x)$ and evaluating $f'(1)$. Show your work.

(c) Repeat part (b), but at $x = 0$. What issue do you run into?

### Solutions

$$f(x) = x(48 - x^2) = 48x - x^3$$

#### Question 2a

To maximize via adequality, we need to:
1. Set $f(a) = f(a + e)$
2. Divide both sides by $e$
3. Set any remaining $e$s to 0
4. Solve for $a$

$$f(a) = f(a+e) \:\:\: (\text{Step 1}) \\
48a - a^3 = 48(a+e) - (a+e)^3 \\
48a - a^3 = 48a + 48e - a^3 - 3a^2e - 3ae^2 - e^3 \\
0 = 48e - 3a^2e - 3ae^2 - e^3 \\
0 = 48 - 3a^2 - 3ae - e^2 \:\:\: (\text{Step 2}) \\
0 = 48 - 3a^2 \:\:\: (\text{Step 3}) \\
a^2 = 16 \implies a = \pm 4 \:\:\: (\text{Step 4})
$$

Note that $f(4) = 4(32) > f(-4) = -4(32)$, so $f(x)$ has a local maximum at $x = 4$. It is not a global maximum since cubic polynomials approach both positive and negative infinity.

#### Question 2b

To find the slope of the tangent line of $f(x) = 48x-x^3$ at $x = 1$, we need to

1. Set $\frac{f(1)}{t} = \frac{f(1+e)}{t+e}$
2. Divide both sides by $e$
3. Set any remaining $e$s to 0
4. Solve for $t$
5. Calculate the slope as $\frac{f(1)}{t}$

$$
\frac{47}{t} = \frac{48 + 48e - 1 - 3e - 3e^2 - e^3}{t + e} \:\:\:(\text{Step 1}) \\ 47t + 47e = 48t + 48et - t - 3et - 3e^2t - e^3t \\
0 = -47e + 48et - 3et - 3e^2t - e^3t \\
0 = -47 + 48t - 3t - 3et - e^2t \:\:\:(\text{Step 2}) \\
0 = -47 + 45t \:\:\:(\text{Step 3}) \\
t = \frac{47}{45} \:\:\:(\text{Step 4})
$$

Then, the slope at $x = 1$ is $$\frac{f(1)}{\frac{47}{45}} = \frac{47}{\frac{47}{45}} = 47 \cdot \frac{45}{47} = \boxed{45}$$

To verify, we can check $f'(x) = 48 - 3x^2$ at $x = 1$, which gives $f'(1) = 48 - 3 = 45$.

#### Question 2c

Since $f(0) = 0$, Step 1 gives us

$$\frac{0}{t} = \frac{48e - e^3}{t+e}$$

Here, there is no solution for $t$. If there were a solution for $t$, the slope $\frac{0}{t}$ would appear to be 0. However, the real slope of the tangent line of $f(x)$ at $x = 0$ is $f'(0) = 48$; the method of adequality fails here.


## Question 3

Show that Fermat's method for finding the slope of the tangent line is equivalent to

$$\text{slope} = \frac{f(x+e) - f(x)}{e}$$

Does the _quotient_ on the right look familiar – if so, what does it resemble?

Hopefully, by answering this question, it is slightly more clear why Fermat's method of adequality involves dividing by $e$ and then setting $e$ to 0.

### Solutions

Recall, the slope of the tangent line is given by $\frac{f(x)}{t}$.

$$\frac{f(x)}{t} = \frac{f(x+e)}{t + e} \\
(t+e) f(x) = t f(x+e) \\
t f(x) + ef(x) = t f(x+e) \\
e f(x) = t \left( f(x+e) - f(x) \right) \\
\frac{f(x)}{t} = \boxed{\frac{f(x+e) - f(x)}{e}}
$$

This quotient resembles the formal definition of a derivative, which states

$$f'(x) = \lim_{h\rightarrow0} \frac{f(x+h) - f(x)}{h}$$

Fermat's method of adequality involves dividing by $e$ and setting $e=0$, which resembles dividing by $h=0$ and then taking the limit as $h \rightarrow 0$.



## Question 4

One aggregation strategy that was used centuries ago involved picking the number halfway between the smallest and largest numbers in a dataset. In today's terminology, we'd call this number the "midrange".

$$\text{Midrange} = \frac{\text{Max}(x) + \text{Min}(x)}{2}$$

For each of the following conditions, specify a set of 5 unique numbers that satisfies the given condition. In each case, compute the midrange and mean and show that the condition holds. Afterwards, give a one sentence answer to the question, "for what kinds of datasets is the midrange close to the mean?"

(a) Midrange = mean

(b) Midrange > mean

(c) Midrange < mean

### Solutions

Remember that the question asked for all example datasets to consist of **unique** values.

#### Question 4a

An example dataset where the midrange is equal to the mean is

$$1, 2, 3, 4, 5$$

The midrange is $\frac{1+5}{2} = 3$, and the mean is $\frac{1 + 2 + 3 + 4 + 5}{5} = 3$.

#### Question 4b

An example dataset where the midrange is greater than the mean is

$$1, 2, 3, 4, 11$$

The midrange is $\frac{1+11}{2} = 6$, and the mean is $\frac{1+2+3+4+11}{5} = 4.2$.

#### Question 4c

An example dataset where the midrange is less than the mean is

$$1, 8, 9, 10, 11$$

The midrange is $\frac{1+11}{2} = 6$, and the mean is $\frac{1+8+9+10+11}{5} = 7.8$.

The midrange is close to the mean in datasets that are symmetric, with equally many large and small values.
